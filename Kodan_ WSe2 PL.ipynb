Plotting data for the **Photoluminescence spectroscopy** of $WSe_2$

```python
import pandas as pd
import numpy as np
import plotly.express as px



df = pd.read_excel('PL Data_Flake 1.xlsx')
df.head()

fig = px.line(df, x='wavelength', y='intensity',labels={'wavelength':'Wavelength (nm)', 'intensity':'Intensity (a.u.)'})

fig.update_layout(
    title={
        'text': "$W Se_2 PL \ (Flake \ 1)$",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

fig.update_layout(
    font_family="Times New Roman",
    font_color="black",
    title_font_family="Times New Roman",
    title_font_color="black",
    font_size=15,
title_font_size = 20)

fig.show()
```

```python
df2 = pd.read_excel('PL Data_Flake 2.xlsx')
df2.head()

fig = px.line(df2, x='Wavelength', y='Intensity',labels={'Wavelength':'Wavelength(nm)', 'Intensity':'Intensity(a.u.)'})


fig.update_layout(
    title={
        'text': "$W Se_2 PL\ Flake $",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

fig.update_layout(
    font_family="Times New Roman",
    font_color="black",
    title_font_family="Times New Roman",
    title_font_color="black",
   font_size=15,
title_font_size = 20)

fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(
    go.Scatter(x=df['wavelength'],y=df['intensity'],
                        name='$1\  Layer \ WSe_2  $'),
             )

fig.add_trace(go.Scatter(x=df2['Wavelength'],y=df2['Intensity']*5,
                        name='$4\  Layer\  WSe_2 (normalized) ///$')
             )

fig.update_layout(
    title={
        'text': "$W Se_2 PL$",
        'y':0.84,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

fig.update_layout(
    font_family="Times New Roman",
    font_color="black",
    title_font_family="Times New Roman",
    title_font_color="black",
    xaxis_title='Wavelength(nm)',
    yaxis_title="Intensity(a.u.)",
       font_size=15,
title_font_size = 20
)

fig.update_layout(legend=dict(
    yanchor="top",
    y=0.99,
    xanchor="left",
    x=0.5
))

fig.update_yaxes(range=[-20, 6000])
fig.show()
```
